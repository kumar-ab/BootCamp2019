```python
import numpy as np
from matplotlib import pyplot as plt
```

## Problem 1:

```python
# Function to create n x n array from random normal distribution sample
def normal_array(num):
    arr_n = np.random.normal(size=(num, num))
    arr_mean = np.mean(arr_n, axis=1)
    v = np.var(arr_mean)
    return v

# Function to call the above functions with different values of n
def fn_call():
    n = np.arange(100, 1100, 100)
    var_arr = np.zeros(len(n))
    for i, nn in enumerate(n):
        var_arr[i] = normal_array(nn)
    return var_arr

# Call the function and plot the result
arr_res = fn_call()
plt.plot(arr_res)
plt.show()
```

## Problem 2:

```python
# define the function
def trig_plot():
    plt.ion()
    x = np.linspace(-2*np.pi, 2*np.pi, 50)
    plt.plot(x, np.sin(x), label="sin(x)")
    plt.plot(x, np.cos(x), label="cos(x)")
    plt.plot(x, np.arctan(x), label="arctan(x)")
    plt.legend(loc="lower right")
    plt.xlabel("x")
    plt.ylabel("sin/cos/arctan")

# Call the function
trig_plot()
```

## Problem 3:

```python
# Function to plot f(x)
def fn_plot():
    x1 = np.linspace(-2, 1, 50)
    x2 = np.linspace(1, 6, 50)
    y1 = 1/(x1-1)
    y2 = 1/(x2-1)
    plt.plot(x1, y1, 'm--', linewidth=4)
    plt.plot(x2, y2, 'm--', linewidth=4)
    plt.xlim(-2, 6)
    plt.ylim(-6, 6)
# Call function
fn_plot()
```

## Problem 4:

```python
# define the function
def sub_plot():
    x = np.linspace(0, 2*np.pi, 50)
    
    # Plot 1: sin(x)
    plt.subplot(2,2,1)
    plt.plot(x, np.sin(x), 'g-', lw=2)
    plt.title("sin(x)")
    # Plot 2: sin(2x)
    plt.subplot(2,2,2)
    plt.plot(x, np.sin(2*x), 'r--', lw=2)
    plt.title("sin(2x)")
    # Plot 3: 2sin(x)
    plt.subplot(2,2,3)
    plt.plot(x, 2* np.sin(x), 'b--', lw=2)
    plt.title("2sin(x)")
    # Plot 4: 2sin(2x)
    plt.subplot(2,2,4)
    plt.plot(x, 2*np.sin(2*x), 'm:', lw=2)
    plt.title("2sin(x)")
    
    plt.axis([0, 2*np.pi, -2, 2])
    plt.suptitle("various sin functions", fontsize=18)

# Call the function
sub_plot()
```

## Problem 5:

```python
def data_vis():
    fars = np.load("FARS.npy")
    x1 = fars[:,2]
    y1 = fars[:,1]
    x2 = fars[:,0]
    # Plot 1: Scatter plot of Longitudes vs Lattitudes
    plt.subplot(1,2,1)
    plt.plot(x1, y1, 'k')
    plt.xlabel("Lattitudes")
    plt.ylabel("Longitudes")
    plt.axis("equal")
    # Plot 2: histogram of hours
    plt.subplot(1,2,2)
    plt.hist(x2, bins=np.arange(0, 23))
    plt.xlabel("Hour of the day")
    plt.xlim(0,23)
# Call function
data_vis()
```
