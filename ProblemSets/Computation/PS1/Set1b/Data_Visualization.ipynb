```python
import numpy as np
from scipy.special import binom
from scipy import stats
from scipy.optimize import minimize
from matplotlib import pyplot as plt
```

## Problem 2:
### Plotting the Bernstein basis polynomials
\begin{equation}
  b_{v,n}(x) = {n\choose v} x^{v}(1-x)^{n-v}, \qquad v = 0, 1,..., n
\end{equation}

```python
x = np.linspace(0, 1, 50)
# Call the function using x
i = 1
for n in range(4):
    for v in range(4):
        if v > n:
            break
        else:
            c = binom(n,v)
            b = c * x**v * (1-x)**(n-v)
            plt.subplot(5,2,i)
            plt.plot(x, b)
            c_title = "(v,n) = (" + str(v) + "," + str(n) + ")"
            plt.title(c_title, y=0.5)
            plt.xlim(0,1)
            plt.ylim(0,1)
            i=i+1
```

## Problem 3:

```python
mlb = np.load("MLB.npy")
ht = mlb[:,0]
wt = mlb[:,1]
age = mlb[:,2]
# Regression with ht as independent variable & wt as dependent
(gr, incpt, r_value, p_value, std_err) = stats.linregress(ht,wt)
# Plots for data visualisation
plt.subplot(2,1,1) # age v/s weight
plt.scatter(age, wt, s=5, c='r')
plt.xlabel("Age in years")
plt.ylabel("Weight in pounds")
plt.xlim(15,50)
plt.ylim(100,300)
plt.subplot(2,1,2) # age v/s height
plt.scatter(age, ht, s=10, c='m')
plt.xlabel("Age in years")
plt.ylabel("Height in inches")
plt.xlim(15,50)
plt.ylim(60,90)
plt.show()
print("p_value of the regression: " + str (p_value))
# height v/s weight
plt.scatter(ht, wt, alpha=0.5, label="actual values")
# The regression line
plt.plot(ht, incpt + gr*ht, c='g', label="Regression line")
plt.xlabel("Height in inches")
plt.ylabel("Weight in pounds")
plt.xlim(65,85)
plt.ylim(100,300)
plt.legend(loc="lower right")
plt.title("Height v Weight Regression")
plt.show()
print("p_value of the regression: " + str (p_value))
```

## Problem 5:
### The Rosenbrock function
\begin{equation}
  f(x,y) = (1-x)^{2} + 100(y-x^{2})^{2}
\end{equation}

```python
# Creating a 2-d domain for x & y
t = np.linspace(-2, 2, 100)
(x, y) = np.meshgrid(t, t)

# Calculating the value of f(x) on the domain
z = (1-x)**2 + 100 * (y-x**2)**2

# Heat map
plt.subplot(2,1,1)
plt.pcolormesh(x, y, z, cmap='inferno')
plt.colorbar()
# Contour map
plt.subplot(2,1,2)
plt.contour(x, y, z, cmap='viridis')
plt.colorbar()
plt.show()
```
