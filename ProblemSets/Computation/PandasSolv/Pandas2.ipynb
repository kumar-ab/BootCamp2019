# Pandas 2: Plotting

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pydataset import data
```

```python
# Road accident deaths in US states
road_deaths = data("road")
road_deaths.index.name = "states"
# The United States personal expenditures (in billions of dollars)
expenditure = data("USPersonalExpenditure")
expenditure = expenditure.transpose()
expenditure.index.name = "Year"
```

```python
# Plot all the data for road deaths
road_deaths[["deaths", "drivers"]].plot(linewidth=1, marker='o')
plt.xticks(list(range(len(road_deaths.index))), road_deaths.index, rotation=90)
plt.show()
```

```python
# Plot all the data for expenditure
expenditure.plot(linewidth=1, marker='o')
plt.xticks(list(range(len(expenditure.index))), expenditure.index)
plt.show()
```
